```python
position_cols = ['lw', 'rw', 'ls', 'st', 'rs', 'lf', 'cf', 'rf', 'lam', 'cam', 'ram', 'lcm', 'cm', 'rcm', 'lm', 'rm', 'ldm', 'cdm', 'rdm', 'lwb', 'rwb', 'lb', 'rb', 'lcb', 'cb', 'rcb']

def convert_position_rating(rating):
    if pd.isna(rating):
        return None
    else:
        parts = rating.split('+')
        return int(parts[0]) + int(parts[1])

data[position_cols] = data[position_cols].applymap(convert_position_rating)
data[position_cols] = data[position_cols].fillna(0)

data['wing'] = data[['lw', 'rw']].mean(axis=1)
data['forward'] = data[['ls', 'st', 'rs', 'lf', 'cf', 'rf']].mean(axis=1)
data['attacking_midfield'] = data[['lam', 'cam', 'ram']].mean(axis=1)
data['midfield'] = data[['lcm', 'cm', 'rcm']].mean(axis=1)
data['wide_midfield'] = data[['lm', 'rm']].mean(axis=1)
data['defensive_midfield'] = data[['ldm', 'cdm', 'rdm']].mean(axis=1)
data['full_back'] = data[['lwb', 'rwb', 'lb', 'rb']].mean(axis=1)
data['defense'] = data[['lcb', 'cb', 'rcb']].mean(axis=1)

data.drop(columns=position_cols, inplace=True)

print(data.head())
```

### Which type of offensive players tends to get paid the most: the striker, the right-winger, or the left-winger?

```python
offensive_positions = ['ST', 'RW', 'LW']
offensive_players = data[data['player_positions'].apply(lambda x: any(pos in x for pos in offensive_positions))]

offensive_players['primary_position'] = offensive_players['player_positions'].str.split(',').str[0]

offensive_players = offensive_players[offensive_players['primary_position'].isin(offensive_positions)]

average_wages = offensive_players.groupby('primary_position')['wage_eur'].mean().sort_values(ascending=False)
```
